## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

df["subtotal"] = df.unit_price * df.qty
df[["unit_price", "qty", "subtotal"]].head()

```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
def shipping(row):
    cost = 0
    num = row['unit_weight'] * row['qty']
    if num < 50:
        cost = num * 10
    if num > 50:
        cost = num * 7
    return cost
df['shipping_price'] = df.apply(shipping, axis=1)
df[["unit_weight", "qty", "shipping_price"]].head()
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

df["line_cost"] = ((df.unit_cost * df.qty) + df.shipping_price).round(2)
df.head(3)
```

```python
# Create a column for the profit of
# each line using line cost and line price

df["line_profit"] = (df.line_price - df.line_cost).round(2)
df.head(3)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
for client in top_clients:
    total = df.loc[df["client_id"] == client, 'line_price'].sum()
    print(f"{client}: ${total.round(2)}")
    
    


```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.
top_clients_list = []
for client in top_clients:
    qty = df.loc[df["client_id"] == client, 'qty'].sum()
    shipping = df.loc[df["client_id"] == client, 'shipping_price'].sum().round(2)
    line_price = df.loc[df["client_id"] == client, 'line_price'].sum().round(2)
    line_cost = df.loc[df["client_id"] == client, 'line_cost'].sum().round(2)
    line_profit = df.loc[df["client_id"] == client, 'line_profit'].sum().round(2)
    values = {
        'client_id' : client,
        'qty' : qty,
        'shipping_price' : shipping,
        'line_price' : line_price,
        'line_cost' : line_cost,
        'line_profit' : line_profit
    }
    top_clients_list.append(values)

top_client_table_df = pd.DataFrame(top_clients_list)
top_client_table_df.sort_values('line_profit', ascending=False)

                        






```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.
renamed_df = top_client_table_df.rename(columns={"client_id":"Client ID", 
                                          "qty":"Units", 
                                          "shipping_price": "Shipping", 
                                          "line_price": "Total Revenue",
                                           "line_cost": "Total Cost",
                                            "line_profit": "Total Profit"})


millions_of_dollars = renamed_df["Total Profit"]/1000000
renamed_df["Total Profit"] = millions_of_dollars.round(2)
renamed_df["Total Profit"] = "$" + renamed_df["Total Profit"].astype(str) + "M"

millions_of_dollars = renamed_df["Total Cost"]/1000000
renamed_df["Total Cost"] = millions_of_dollars.round(2)
renamed_df["Total Cost"] = "$" + renamed_df["Total Cost"].astype(str) + "M"

millions_of_dollars = renamed_df["Total Revenue"]/1000000
renamed_df["Total Revenue"] = millions_of_dollars.round(2)
renamed_df["Total Revenue"] = "$" + renamed_df["Total Revenue"].astype(str) + "M"

millions_of_dollars = renamed_df["Shipping"]/1000000
renamed_df["Shipping"] = millions_of_dollars.round(2)
renamed_df["Shipping"] = "$" + renamed_df["Shipping"].astype(str) + "M"

renamed_df.head()

```

```python
# Sort the updated data by "Total Profit" form highest to lowest
renamed_df.sort_values('Total Profit', ascending=False)
```
